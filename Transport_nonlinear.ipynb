```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
length=10
a=1
Currant=0.7
x_steps_amount=100
t_steps_amount=100
duration=Currant*length*(t_steps_amount-1)/(a*(x_steps_amount-1))
FirstOrder=np.zeros((x_steps_amount,t_steps_amount))
Model=np.zeros((x_steps_amount,t_steps_amount))
TVD=np.zeros((x_steps_amount,t_steps_amount))
start_value=0
start_impulse=1
```

```python
x_step=length/(x_steps_amount-1)
t_step=duration/(t_steps_amount-1)
```

```python
Currant=t_step/x_step
print(Currant)
```

```python
FirstOrder[:,0]=Model[:,0]=TVD[:,0]=start_value
FirstOrder[0,:]=Model[0,:]=TVD[0,:]=start_impulse
FirstOrder[0][0]=Model[0][0]=TVD[0][0]=start_impulse
```

```python
#count trianglep
for t in range(0,t_steps_amount-1):
    for x in range(1,x_steps_amount):
        #print(Z[x][t]-Currant*(Z[x][t]-Z[x-1][t]))
        FirstOrder[x][t+1]=FirstOrder[x][t]-Currant*((FirstOrder[x][t]**2)/2-(FirstOrder[x-1][t]**2)/2)
```

```python
#count TVD
for t in range(0,t_steps_amount-1):
    for x in range(1,x_steps_amount-1):
        e=1e-7
        r=(TVD[x][t]-TVD[x-1][t]+e)/(TVD[x+1][t]-TVD[x][t]+e)
        #print(r)
        fi=0;
        if r>1:
            fi=min(r,2)
        elif r>0:
            fi=min(2*r,1)
        fplus=0.5*fi*Currant*(1-Currant)*0.5*(TVD[x+1][t]**2-TVD[x][t]**2)
        fminus=0.5*fi*Currant*(1-Currant)*0.5*(TVD[x][t]**2-TVD[x-1][t]**2)
        TVD[x][t+1]=TVD[x][t]-Currant*0.5*(TVD[x][t]**2-TVD[x-1][t]**2)-(fplus-fminus)
```

```python
#count model
for x in range(0,x_steps_amount-1):
    for t in range(1,t_steps_amount-1):
        speed=(start_impulse+start_value)/2
        if(x*x_step<=speed*t*t_step):
            Model[x][t]=start_impulse
```

```python
def perestroyka(q,i):
    a=[]
    for f in q.tolist():
        a.append(f[i])
    a=np.array(a)
    return a
```

```python
import matplotlib.pyplot as plt
plt.close()
plt.figure(figsize=(40,10))
ax1=plt.subplot(1,1,1)
ax1.set_xlim(0,x_steps_amount*x_step)
ax1.set_ylim(0,1.1)
line0, =ax1.plot(np.linspace(0,x_step*x_steps_amount,x_steps_amount),perestroyka(FirstOrder,50),c='black')
line1, =ax1.plot(np.linspace(0,x_step*x_steps_amount,x_steps_amount),perestroyka(Model,50),c='blue')
line2, =ax1.plot(np.linspace(0,x_step*x_steps_amount,x_steps_amount),perestroyka(TVD,50),c='red')
plt.savefig("nonlinear.png")
plt.show()
```
