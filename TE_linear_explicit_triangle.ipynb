```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
duration=200
length=2.5
a=0.015
x_steps_amount=100
t_steps_amount=101
Z=np.zeros((x_steps_amount,t_steps_amount))

start_value=0
start_impulse=1
```

```python
x_step=length/(x_steps_amount-1)
t_step=duration/(t_steps_amount-1)
```

```python
Z[:,0]=start_value
Z[0,:]=start_impulse
Z[0][0]=1
```

```python
Currant=a*t_step/x_step
if Currant >1 :
    print("Change Currant")
Currant=1
```

```python
for t in range(0,t_steps_amount-1):
    for x in range(1,x_steps_amount):
        #print(Z[x][t]-Currant*(Z[x][t]-Z[x-1][t]))
        Z[x][t+1]=Z[x][t]-Currant*(Z[x][t]-Z[x-1][t])
```

```python
for t in range(0,54,3):
    plt.subplot(10,2,t/3+1)
    plt.plot(Z[:,t],c='black')
    plt.subplots_adjust(top=3,bottom=0.001,hspace =1)
```
